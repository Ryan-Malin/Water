Collecting predictions from Riley's machine learning models

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('water_potability.csv')
df.head()
```

```python
df.isna().sum()
```

```python
# add index identifier column
df = df.reset_index().rename(columns = {"index": "id"})
df
```

```python
df = df.dropna(how='any')
df.shape
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
scaler = MinMaxScaler()

# scale all except index [0] and potability [-1]
array = scaler.fit_transform(df[df.columns[1:-1]].to_numpy())

# new dataframe with scaled measures
df_scaled = df.copy().set_index("id")
i = 0
for c in df.columns[1:-1]:
    df_scaled[c] = array.transpose()[i]
    i += 1

df_scaled
```

```python
df_scaled.describe()
```

```python
# check that the index numbers are correct
list(df_scaled.index) == list(df["id"])
```

```python
# Preparing variables
X = df_scaled.drop(['Potability'], axis = 1)
y = df_scaled['Potability']
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pca = PCA(n_components=0.90)

water_pca = pca.fit_transform(X)
```

```python
pca_df = pd.DataFrame(
        data=water_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8']
)
pca_df["id"] = df_scaled.index

pca_df = pca_df.join(y, on = "id")
pca_df
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (6,6))
sb.scatterplot(data = pca_df , x = 'PC1',y = 'PC5' , hue = 'Potability' , s = 60)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=1111, stratify=y) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,roc_curve,accuracy_score, classification_report
```

```python
k = range(1,20,1)
testing_accuracy = []
training_accuracy = []

score = 0

# knn
for i in k:
    knn= KNeighborsClassifier(n_neighbors=i, metric='minkowski', p=2 )  
    knn.fit(X_train, y_train)  
    y_pred_train = knn.predict(X_train)
    training_accuracy.append(accuracy_score(y_train, y_pred_train))
    
    y_pred_test = knn.predict(X_test)
    acc_score = accuracy_score(y_test,y_pred_test)
    testing_accuracy.append(acc_score)
    
    print(i, acc_score)
    
    if score < acc_score:
        score = acc_score
        best_k = i
        
print('Best Accuracy Score for PCA from KNN Model', score, 'Best K-Score', best_k)
```

```python
knn = KNeighborsClassifier(n_neighbors=17, metric='minkowski', p=2 )
knn.fit(X_train, y_train)  
y_pred_train_knn = knn.predict(X_train)
y_pred_test_knn = knn.predict(X_test)
print(accuracy_score(y_test, y_pred_test_knn))
```

```python
# knn predictions
knn_train = pd.DataFrame(y_pred_train_knn, index = y_train.index, columns = ["knn"])
knn_test = pd.DataFrame(y_pred_test_knn, index = y_test.index, columns = ["knn"])

knn_df = pd.concat([knn_train, knn_test], axis = 0)
knn_df
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='rbf')

svm = svm.fit(X_train, y_train)

y_train_pred_svm = svm.predict(X_train)
y_test_pred_svm =svm.predict(X_test)
acc = accuracy_score(y_test,y_test_pred_svm)

print('Best Accuracy Score for SVM', acc, 'model is able to succesfully predict on waterpotability with a value of ', acc)
```

```python
svm_train = pd.DataFrame(y_train_pred_svm, index = y_train.index, columns = ["svm"])
svm_test = pd.DataFrame(y_test_pred_svm, index = y_test.index, columns = ["svm"])

svm_df = pd.concat([svm_train, svm_test], axis = 0)
svm_df
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clasification = RandomForestClassifier()
clasification.fit(X,y)

variables = X.columns
importance = pd.DataFrame()

importance['Features'] = variables
importance['Importance'] = clasification.feature_importances_
importance = importance.sort_values(by=['Importance'],ascending=True)

labels = importance['Features']
values = importance['Importance']

plt.figure(figsize=(10,10))
plt.pie(values, labels=labels)
plt.show()
```

```python
clf = RandomForestClassifier(random_state=1, n_estimators=6, bootstrap=True, max_samples=1000).fit(X_train, y_train)

y_train_pred_rf = clf.predict(X_train)
y_test_pred_rf = clf.predict(X_test)

print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
features = clf.feature_importances_
print(features)
plt.bar(x = range(len(features)), height=features)
plt.show()
```

```python
rf_train = pd.DataFrame(y_train_pred_rf, index = y_train.index, columns = ["rf"])
rf_test = pd.DataFrame(y_test_pred_rf, index = y_test.index, columns = ["rf"])

rf_df = pd.concat([rf_train, rf_test], axis = 0)
rf_df
```

```python
# Linear Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
