```python
water['ph_rank'] = water[water['ph'].notnull()].groupby('ph', sort=True).ngroup()+1
water['ph_rank'] = water['ph_rank'].fillna('')
```

```python
water['Hardness_rank'] = water[water['Hardness'].notnull()].groupby('Hardness', sort=True).ngroup()+1
water['Hardness_rank'] = water['Hardness'].fillna('')
```

```python
water['Solids_rank'] = water[water['Solids'].notnull()].groupby('Solids', sort=True).ngroup()+1
water['Solids_rank'] = water['Solids_rank'].fillna('')
```

```python
water['Chloramines_rank'] = water[water['Chloramines'].notnull()].groupby('Chloramines', sort=True).ngroup()+1
water['Chloramines_rank'] = water['Chloramines_rank'].fillna('')
```

```python
water['Sulfate_rank'] = water[water['Sulfate'].notnull()].groupby('Sulfate', sort=True).ngroup()+1
water['Sulfate_rank'] = water['Sulfate_rank'].fillna('')
```

```python
water['Conductivity_rank'] = water[water['Conductivity'].notnull()].groupby('Conductivity', sort=True).ngroup()+1
water['Conductivity_rank'] = water['Conductivity_rank'].fillna('')
```

```python
water['Organic_carbon_rank'] = water[water['Organic_carbon'].notnull()].groupby('Organic_carbon', sort=True).ngroup()+1
water['Organic_carbon_rank'] = water['Organic_carbon_rank'].fillna('')
```

```python
water['Trihalomethanes_rank'] = water[water['Trihalomethanes'].notnull()].groupby('Trihalomethanes', sort=True).ngroup()+1
water['Trihalomethanes_rank'] = water['Trihalomethanes_rank'].fillna('')
```

```python
water['Turbidity_rank'] = water[water['Turbidity'].notnull()].groupby('Turbidity', sort=True).ngroup()+1
water['Turbidity_rank'] = water['Turbidity_rank'].fillna('')
```

```python
def check(x):
    if x['Hardness_Safe'] == True & x['ph_safe'] == True & x['Conductivity_safe'] == True & x['Trihalomethanes_safe'] == True & x['Turbidity_safe'] == True:
        return True
    else:
        return False
water4['Potable_WHO_Standards'] = water4.apply(check, axis=1)
```

```python
def potable(y):
    if y['Potability'] == 1:
        return True
    else:
        return False
water4['Potability'] = water4.apply(potable, axis=1)
```
