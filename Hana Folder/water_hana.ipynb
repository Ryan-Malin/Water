```python
import pandas as pd
```

```python
water_data_to_load ="water_potability.csv"
water_data = pd.read_csv(water_data_to_load)
```

```python
water_data.dropna()
water_data.reset_index()
```

```python
water_data['ph_rank'] = water_data[water_data['ph'].notnull()].groupby('ph', sort=True).ngroup()+1
water_data['ph_rank'] = water_data['ph_rank'].fillna('')

water_data['Hardness_rank'] = water_data[water_data['Hardness'].notnull()].groupby('Hardness', sort=True).ngroup()+1
water_data['Hardness_rank'] = water_data['Hardness'].fillna('')

water_data['Solids_rank'] = water_data[water_data['Solids'].notnull()].groupby('Solids', sort=True).ngroup()+1
water_data['Solids_rank'] = water_data['Solids_rank'].fillna('')

water_data['Chloramines_rank'] = water_data[water_data['Chloramines'].notnull()].groupby('Chloramines', sort=True).ngroup()+1
water_data['Chloramines_rank'] = water_data['Chloramines_rank'].fillna('')

water_data['Sulfate_rank'] = water_data[water_data['Sulfate'].notnull()].groupby('Sulfate', sort=True).ngroup()+1
water_data['Sulfate_rank'] = water_data['Sulfate_rank'].fillna('')

water_data['Conductivity_rank'] = water_data[water_data['Conductivity'].notnull()].groupby('Conductivity', sort=True).ngroup()+1
water_data['Conductivity_rank'] = water_data['Conductivity_rank'].fillna('')

water_data['Organic_carbon_rank'] = water_data[water_data['Organic_carbon'].notnull()].groupby('Organic_carbon', sort=True).ngroup()+1
water_data['Organic_carbon_rank'] = water_data['Organic_carbon_rank'].fillna('')

water_data['Trihalomethanes_rank'] = water_data[water_data['Trihalomethanes'].notnull()].groupby('Trihalomethanes', sort=True).ngroup()+1
water_data['Trihalomethanes_rank'] = water_data['Trihalomethanes_rank'].fillna('')

water_data['Turbidity_rank'] = water_data[water_data['Turbidity'].notnull()].groupby('Turbidity', sort=True).ngroup()+1
water_data['Turbidity_rank'] = water_data['Turbidity_rank'].fillna('')
```

```python
water_data
```

```python
water_data["Average_Rank"] = water_data[['ph_rank', 'Hardness_rank', 'Solids_rank',
       'Chloramines_rank', 'Sulfate_rank', 'Conductivity_rank',
       'Organic_carbon_rank', 'Trihalomethanes_rank', 'Turbidity_rank']].mean(numeric_only=True, axis=1)
```

```python
water_data = water_data.sort_values("Average_Rank", ascending=True).reset_index(drop=True)
water_data['Average_Rank_2'] = water_data.index
```

```python
water_data[(water_data['Hardness']>=60) & (water_data['Hardness']<=200) & (water_data['Hardness'] != "NaN")]
water_data['Hardness_Safe'] = (water_data['Hardness']>=60) & (water_data['Hardness']<200) & (water_data['Hardness']!='NaN')
water_data['ph_safe'] = (water_data['ph']>=6.5) & (water_data['ph']<8.5) & (water_data['ph']!='NaN')
water_data['Solids_safe'] = (water_data['Solids']<=1000) & (water_data['Solids']!='NaN')
water_data['Chloramines_safe'] = (water_data['Chloramines']<=4) & (water_data['Chloramines']!='NaN')
water_data['Sulfate_safe'] = (water_data['Sulfate']<250) & (water_data['Sulfate']!='NaN')
water_data['Conductivity_safe'] = (water_data['Conductivity']<400) & (water_data['Conductivity']!='NaN')
water_data['Organic_carbon_safe'] = (water_data['Organic_carbon']<2) & (water_data['Organic_carbon']!='NaN')
water_data['Trihalomethanes_safe'] = (water_data['Trihalomethanes']<80) & (water_data['Trihalomethanes']!='NaN')
water_data['Turbidity_safe'] = (water_data['Turbidity']<5) & (water_data['Turbidity']!='NaN')


water_data
```

```python

water_data["Hardness_Safe_qty"] = water_data["Hardness_Safe"].sum()
water_data["ph_saf_datae_qty"] = water_data["ph_safe"].sum()
water_data["Solids_safe_qty"] = water_data["Solids_safe"].sum()
water_data["Chloramines_safe_qty"] = water_data["Chloramines_safe"].sum()
water_data["Sulfate_safe_qty"] = water_data["Sulfate_safe"].sum()
water_data["Conductivity_safe_qty"] = water_data["Conductivity_safe"].sum()
water_data["Organic_carbon_safe_qty"] = water_data["Organic_carbon_safe"].sum()
water_data["Trihalomethanes_safe_qty"] = water_data["Trihalomethanes_safe"].sum()
water_data["Turbidity_safe_qty"] = water_data["Turbidity_safe"].sum()
```

```python
water_data["Hardness_Safe_pct"] = (water_data["Hardness_Safe"].sum()/32.76).round(2)
water_data["ph_safe_pct"] = (water_data["ph_safe"].sum()/32.76).round(2)
water_data["Solids_safe_pct"] = (water_data["Solids_safe"].sum()/32.76).round(2)
water_data["Chloramines_safe_pct"] = (water_data["Chloramines_safe"].sum()/32.76).round(2)
water_data["Sulfate_safe_pct"] = (water_data["Sulfate_safe"].sum()/32.76).round(2)
water_data["Conductivity_safe_pct"] = (water_data["Conductivity_safe"].sum()/32.76).round(2)
water_data["Organic_carbon_safe_pct"] = (water_data["Organic_carbon_safe"].sum()/32.76).round(2)
water_data["Trihalomethanes_safe_pct"] = (water_data["Trihalomethanes_safe"].sum()/32.76).round(2)
water_data["Turbidity_safe_pct"] = (water_data["Turbidity_safe"].sum()/32.76).round(2)
```

```python
# Creating  Turbidity threshold, should be around 5, so around 4.5 and 5.5
passing_turb= water_data[water_data['Turbidity']<=5.5]
passing_turb = passing_turb[water_data['Turbidity']>=4.5]
water_data = passing_turb
water_data
```
